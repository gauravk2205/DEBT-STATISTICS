```python
import pandas as pd

# Load the dataset
df = pd.read_csv("IDSData.csv")

# Display first few rows
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Drop unwanted columns
df.drop(columns=["Unnamed: 59"], errors="ignore", inplace=True)
```

```python
df.rename(columns={"Country Name": "country", "Indicator Name": "indicator_name"}, inplace=True)
```

```python
year_columns = [col for col in df.columns if col.isdigit()]
print("Year Columns:", year_columns)
```

```python
df_melted = df.melt(id_vars=["country", "indicator_name"],
                     value_vars=year_columns,
                     var_name="year",
                     value_name="debt")

# Convert "year" to integer
df_melted["year"] = df_melted["year"].astype(int)

# Drop NaN values in debt column
df_melted.dropna(inplace=True)

# Display cleaned data
df_melted.head()
```

```python
distinct_countries = df_melted["country"].unique()
print("Distinct Countries:", distinct_countries)
```

```python
distinct_indicators = df_melted["indicator_name"].unique()
print("Distinct Debt Indicators:", distinct_indicators)
```

```python
total_debt = df_melted["debt"].sum()
print("Total Debt:", total_debt)
```

```python
highest_debt = df_melted.groupby("country")["debt"].sum().reset_index()
highest_debt = highest_debt.sort_values(by="debt", ascending=False).head(1)
print(highest_debt)
```

```python
avg_debt = df_melted.groupby("indicator_name")["debt"].mean().reset_index()
print(avg_debt)
```

```python
principal_repayments = df_melted[df_melted["indicator_name"].str.contains("Principal repayments", na=False)]
highest_principal = principal_repayments.groupby(["country", "indicator_name"])["debt"].max().reset_index()
highest_principal = highest_principal.sort_values(by="debt", ascending=False).head(1)
print(highest_principal)
```

```python
common_indicator = df_melted["indicator_name"].value_counts().reset_index().head(1)
common_indicator.columns = ["indicator_name", "count"]
print(common_indicator)
```

```python
df_melted.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(df_melted["debt"], bins=50, kde=True)
plt.title("Debt Distribution")
plt.show()
```

```python
debt_trend = df_melted.groupby("year")["debt"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=debt_trend, x="year", y="debt", marker="o")
plt.title("Total Debt Over Time")
plt.xlabel("Year")
plt.ylabel("Debt")
plt.grid()
plt.show()
```

```python
def get_highest_debt(df):
    highest_debt = df.groupby("country")["debt"].sum().reset_index()
    highest_debt = highest_debt.sort_values(by="debt", ascending=False).head(1)
    return highest_debt

print(get_highest_debt(df_melted))
```

```python
top_countries = df_melted.groupby("country")["debt"].sum().reset_index()
top_countries = top_countries.sort_values(by="debt", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x="debt", y="country", palette="viridis")
plt.title("Top 10 Countries with Highest Debt")
plt.xlabel("Total Debt")
plt.ylabel("Country")
plt.show()
```

```python
indicator_trend = df_melted[df_melted["indicator_name"] == "Total debt service"]
indicator_trend = indicator_trend.groupby("year")["debt"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=indicator_trend, x="year", y="debt", marker="o", color="red")
plt.title("Total Debt Service Over Time")
plt.xlabel("Year")
plt.ylabel("Debt")
plt.grid()
plt.show()
```
